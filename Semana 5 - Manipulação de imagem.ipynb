```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
foto = io.imread('pic.jpg')
plt.imshow(foto)
```

```python
print(foto.shape)
```

```python
plt.imshow(foto[::-1])
```

```python
plt.imshow(foto[:410, :720])
```

```python
foto_r = foto.copy()
foto_r[:,:,1] = 0
foto_r[:,:,2] = 0
plt.imshow(foto_r)
```

```python
plt.imshow(foto[:,:,0], cmap='Greys')
```

```python
foto_mask = np.where(foto > 100, 255, 0)
plt.imshow(foto_mask)
```
